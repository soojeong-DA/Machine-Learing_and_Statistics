# 이상치(Outlier) 처리
1. 표준편차(standard deviation) 이용
    - 현재 분포의 표준편차 기준 +3 이상이거나 -3 이하인 경우 극단치로 처리
    - 정규분포일 경우 유용

2. 표준점수(z-score) 이용
    - 평균을 0, 표준편차를 1로 맞춘후 표준편차 이용 (+-3 벗어난 경우 이상치로 판정)
    - 정규분포일 경우 유용

3. IQR 방식
    - 75 percentile + IQR * 1.5 이상이거나 25 percentile + IQR * 1.5 이하인 경우 극단치로 처리
    - 정규분포 아닐 경우 robust한 편

4. 기타 방법
    - Rule-based (ex, percentile 95% 이상이면 제거)
    - Binning (연속변인을 카테고리형으로 변환)


→ 표준편차 방법을 원 데이터로 처리시 매우 보수적인 결과 -> **효과가 낮아서 자주 쓰이지 않음**     
→ 일반적으로 **z-score** 자주 활용, 단 정규분포에 효과적      
→ 정규분포가 아닐 경우, **IQR** 고려

````python
# 1. 원 데이터의 표준편차 이용 (6-sigma)
# 각 컬럼별 표준편차에서 +-3 std 벗어난 경우 제외
def std_based_outlier(df):
    for i in range(0, len(df.columns)):
        df = df[~np.abs(df.iloc[:,i] - df.iloc[:,i].mean() > (3*df.iloc[:,i].std()))]
        return df
````

````python
# 더 폭넓은 이상치 제거
# 2. z-score 이용 - 표준점수로 변환 후 +-3 std 벗어나는 경우 제거
df3_zscore = df3[(np.abs(stats.zscore(df3)) < 3).all(axis=1)]   # 모든 컬럼이 +-3 std 벗어나지 않는 것만
df3_zscore.shape
````

````python
df3_zscore.plot(kind='scatter', x='viewCount', y='editCount')
````

````python
# IQR

df3 = df2._get_numeric_data()

for i in range(0, len(df3.columns)):
    q1 = df3.iloc[:,i].quantile(0.25)
    q3 = df3.iloc[:,i].quantile(0.75)
    iqr = q3-q1
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr       

    df3 = df3[(df3.iloc[:,i] >= fence_low) & (df3.iloc[:,i] <= fence_high)]
````

**[How the grid_search module works]**
```python

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# SET default
best_score = 0

# iterataion
for r in ['l1', 'l2']:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        lm = LogisticRegression(penalty = r, C=C)
        scores = cross_val_score(lm, X_train, y_train, cv=5)
        score = np.mean(scores)
        if score > best_score:
            best_score = score
            best_parameters = {'C': C, 'penalty': r}
            
```

## Upsampling & Downsampling for imbalanced data
- target 변수의 그룹별 비율이 불균형인 경우 아래와 같은 방법 고려          
    1. Collect More Data (if possible)
    2. Resampling the Dataset
        - oversampling
            - no information loss, perform better than undersampling
            - overfitting issues (because of duplicates)
        - undersampling
            - help improve run time and storage problems
            - information loss, biased dataset
    3. Generate Synthetic Samples

````python
# target label 1 - oversampling
def oversampling(df):
    
    df_pay_only = df[df['group'] == 1]
    df_pay_only_over = pd.concat([df_pay_only, df_pay_only, df_pay_only], axis=0)
    df_over = pd.concat([df, df_pay_only_over], axis=0)
    
    return df_over
````

### Univariate Selection
- 단변량 특성선택으로 F-value를 이용하여 변수 선택
- 각 변수마다 F-value를 구해 F-value가 큰 변수를 기준으로 변수 선택
    - 그룹내 분산이 작고, 그룹간 분산이 클 경우 F value가 커짐    
    → F value가 크다는 의미는 그룹간 통계적 차이가 크다는 것을 의미

````python
from sklearn.feature_selection import SelectKBest, f_classif

def pipeline_logit_kbest(X_train, y_train):
    # if not classificate 
    #-> regression problem, socre_func=f_regression
    select = SelectKBest(score_func=f_classif)  
    
    pipe = Pipeline([('scaler', StandardScaler()), ('feature_selection', select), ('model', LogisticRegression())])
    
    param_grid = [{
                  'feature_selection__k':[3,5,7,9,11],    # 남길 변수 개수
                  'model__C':[0.001, 0.01, 0.1, 1, 10, 100],
                  'model__penalty':['l1', 'l2']
                    }]
    # grid with pipe
    grid_search = GridSearchCV(pipe, param_grid, scoring='roc_auc',cv=5)
    grid_search.fit(X_train, y_train)
    
    return grid_search
````

````python
from sklearn.ensemble import RandomForestClassifier

def pipeline_tree(X_train, y_train):
    # if not classificate 
    #-> regression problem, socre_func=f_regression
#     select = SelectKBest(score_func=f_classif)  
    
    # tree model - scaling 필요 없음
    pipe = Pipeline([('model', RandomForestClassifier())])
    
    param_grid = [{'model__max_depth':[4,6,8],
                    # n_estimators: The number of trees in the forest.
                   'model__n_estimators':[10,50],
                   'model__min_samples_split':[50,100]
                    }]
    # grid with pipe
    grid_search = GridSearchCV(pipe, param_grid, scoring='roc_auc', cv=5)
    grid_search.fit(X_train, y_train)
    
    return grid_search
````

- Backward 방식 중 하나로, 모든 변수를 다 포함시키고 반복해서 학습을 하면서, 중요하지 않은 변수를 하나씩 제거하는 방식

````python
## RFE (Recursive Feature Elimination)
# short ver.
from sklearn.svm import LinearSVC
from sklearn.feature_selection import RFE

svm = LinearSVC()
rfe = RFE(svm, 1)
rfe = rfe.fit(X_train, y_train)

print(X.columns.tolist())
print(rfe.support_)
print(rfe.ranking_)
````

````python
# pipeline ver.
def pipeline_lm_rfe(X_train, y_train):

    scaler = StandardScaler()
    lm = LogisticRegression()
    select = RFE(lm, 1) 
    
    pipe = Pipeline([('feature_selection', select), 
                     ('scaler', scaler), 
                     ('model', lm)])

    param_grid = [{'model__C': [0.01, 1], 
                  'model__penalty': ['l1', 'l2']
                  }]

    grid_search = GridSearchCV(pipe, param_grid, scoring='roc_auc', cv=2)
    grid_search.fit(X_train, y_train)
    
    return grid_search
````

````python
grid_search_rfe = pipeline_lm_rfe(X_train, y_train)
````

````python
evaluation(grid_search_rfe, X_test, y_test)
````
